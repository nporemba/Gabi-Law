```python
import pandas as pd 
import numpy as np
```

```python
# Get CSV data 
games_13 = pd.read_csv('https://github.com/ixarchakos/nba-games/blob/master/data/nba_games_2013_2014.csv?raw=true') 
games_14 = pd.read_csv('https://github.com/ixarchakos/nba-games/blob/master/data/nba_games_2014_2015.csv?raw=true') 
games_15 = pd.read_csv('https://github.com/ixarchakos/nba-games/blob/master/data/nba_games_2015_2016.csv?raw=true') 
```

```python
# Remove unwanted columns 
keep_cols = ['id', 'date', 'home_team', 'home_points', 'away_team', 'away_points'] 
games_13 = games_13[keep_cols] 
games_14 = games_14[keep_cols] 
games_15 = games_15[keep_cols] 

# Replace team names so that they are consistent between dataframes 
teams = ['Phoenix Suns', 'Chicago Bulls', 'Philadelphia 76ers', 'Orlando Magic', 'Cleveland Cavaliers', 'Milwaukee Bucks', 'Charlotte Hornets', 'LA Lakers', 'Brooklyn Nets', 'Miami Heat'] 
teams_abrv = ['PHX', 'CHI', 'PHI', 'ORL', 'CLE', 'MIL', 'CHA', 'LAL', 'BKN', 'MIA'] 

games_13 = games_13.replace(teams, teams_abrv) 
games_14 = games_14.replace(teams, teams_abrv) 
games_15 = games_15.replace(teams, teams_abrv) 

games_13.head()
```

```python
# Remove observations that do not involve the 10 randomly selected teams 
# df[df.color.isin(['red', 'green', 'blue'])] 
games_13 = games_13[games_13.home_team.isin(teams_abrv)] 
games_13 = games_13[games_13.away_team.isin(teams_abrv)] 

games_14 = games_14[games_14.home_team.isin(teams_abrv)] 
games_14 = games_14[games_14.away_team.isin(teams_abrv)] 

games_15 = games_15[games_15.home_team.isin(teams_abrv)] 
games_15 = games_15[games_15.away_team.isin(teams_abrv)] 

games_13.head()
```

```python
# Add column for win or loss 
# df.assign(new_col = df['col'] * val) 
games_13 = games_13.assign(home_win = games_13['home_points'] > games_13['away_points']) 
games_14 = games_14.assign(home_win = games_14['home_points'] > games_14['away_points']) 
games_15 = games_15.assign(home_win = games_15['home_points'] > games_15['away_points']) 

games_13.head()
```
